```python
import pandas as pd 
import numpy as np 
from dask import dataframe as dd

```

```python
dataset_path = '/Users/tushar/iisc/sem3/TSE/project/Dataset/output/stackoverflow/Posts.csv'
random_state = 33 
```

```python
# using a function so we can track memory usage

def dask_read_and_sample(blocksize, sample_size):
    
    # reading train data
    df_train = dd.read_csv(dataset_path, blocksize=blocksize,dtype={'OwnerUserId': 'float64'})
    
    # let us stratify to get the same number of rows for frauds and non-frauds
    sample_positive = df_train.query('ClosedDate.isnull()').sample(random_state=random_state,frac=sample_size)
    sample_negative = df_train.query('ClosedDate.notnull()')
    
    # concatenate the dataframe
    df_sampled = dd.concat([sample_positive, sample_negative])
    
    return df_sampled.compute()

# executing
df_train = dask_read_and_sample(blocksize=10e6, sample_size=0.02)
```

```python
df_train.to_csv('sample_0.02_33.csv')
```
